```python
sns.displot(resto.total_bill)
```

```python
resto.total_bill.skew()
```

```python
plt.boxplot(resto.total_bill)
```

```python
plt.figure(figsize =(3,3))
plt.rcParams["figure.figsize"] = (10,6)
sns.scatterplot(data = resto , x="total_bill" , y ="tip", hue="sex")
```

```python
plt.scatter(x=resto["total_bill"] , y=resto["tip"])
```

```python
# heatmap works only for the numerical column so before creating heatmap we separate the numerical col
import numpy as np

cat_columns = resto.select_dtypes("category")
num_columns =resto.select_dtypes(np.number)

cat_columns.head()


```

```python
num_columns.head()
```

```python
num_columns.corr()
```

```python
sns.heatmap(num_columns.corr() , annot = True) 
```

```python
corr_matrix = resto.corr(numeric_only=True)
print(corr_matrix)
sns.heatmap(corr_matrix)
```
